```python
#Imports
import dash
import dill
from dash import dcc
from dash import html
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output, State
from dash.exceptions import PreventUpdate
import plotly.graph_objects as go
import  plotly.express as px
import pandas as pd
import base64
#Initialising the dash app
app = dash.Dash(__name__, external_stylesheets = [dbc.themes.BOOTSTRAP])
server = app.server

#Create environment workspace
df_1_16 = pd.read_csv("C:/Users/Nikiforos/Desktop/Black_sea/Dash task_02 Feb 2022/Y2016/Dataset_Y2016_map data_1.csv")
df_2_16 = pd.read_csv("C:/Users/Nikiforos/Desktop/Black_sea/Dash task_02 Feb 2022/Y2016/Dataset_Y2016_map data_2.csv")

df_1_17 = pd.read_csv("C:/Users/Nikiforos/Desktop/Black_sea/Dash task_02 Feb 2022/Y2017/Dataset_Y2017_map data_1.csv")
df_2_17 = pd.read_csv("C:/Users/Nikiforos/Desktop/Black_sea/Dash task_02 Feb 2022/Y2017/Dataset_Y2017_map data_2.csv")


# logo = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2017/Simulated spatial distribution/Negative ESI mz=44.9996 RTI=932.2 rt=13.79.png"
# logo_base64 = base64.b64encode(open(logo, 'rb').read()).decode('ascii')
# src = 'data:image/png;base64,{}'.format(logo_base64)

app.layout = html.Div([
    html.Br([]),
    html.H1("EMBLAS-II: EU/UNDP project Improving Environmental Monitoting in the Black Sea", style = {"text-align": "center"}),
    html.H3("Non-target Screening Data", style = {"text-align": "center"}),

    dbc.Row([
        dbc.Col(
            # This is the Card for the control and the table
            dbc.Card([
                # Control Beginning
                    html.H5("Joint Black Sea Surveys"),
                    html.Br([]),
                    html.H6("Select (JBBS) year"),
                    dcc.Dropdown(
                        id = "year_dropdown", # These IDs are important for the callbacks
                        options = [{'label': "2016", 'value': "2016"},
                                   {'label': "2017", 'value': "2017"}],
                        style = {"padding-left": "24px"},
                        value = "2016" # default value
                    ),
                    html.Br([]),
                    html.H6("Select Source"),
                    dcc.Checklist(
                        id = "source_checklist",
                        options=[
                            {'label': "Danube", 'value': 'Danube'},
                            {'label': 'Dnieper', 'value': 'Dnieper'},
                            {'label': 'Unknown origin', 'value': 'Unknown origin'}
                        ],
                        labelStyle={'display': 'block'},
                        inputStyle={"margin-right": "12px"},
                        value=['Danube', "Dnieper", "Unknown origin"],
                        style = {"padding-left": "24px"},
                ),
                html.Br([]),
                html.H6("Filter Compounds"),
                dbc.Row(
                    [
                        dbc.Label("mz", width=4),
                        dbc.Col(
                            dbc.Input(
                                type="number", id="mz_input"
                            ),
                            width=8,
                        ),
                    ],
                    style={"padding-left": "24px"},
                    className="mb-3",
                ),
                dbc.Row(
                    [
                        dbc.Label("RTI", width=4),
                        dbc.Col(
                            dbc.Input(
                                type="number", id="rti_input"
                            ),
                            width=8,
                        ),
                    ],
                    style={"padding-left": "24px"},
                    className="mb-3",
                ),
                dbc.Row(
                    [
                        dbc.Label("rt", width=4),
                        dbc.Col(
                            dbc.Input(
                                type="number", id="rt_input"
                            ),
                            width=8,
                        ),
                    ],
                    style={"padding-left": "24px"},
                    className="mb-3",
                ),
                html.Br([]),
                html.H6("Select Compound"),
                dcc.Dropdown(
                    id="compound_dropdown",
                    optionHeight=25,
                    style={"padding-left": "24px"},
                ),
            ], style = { "height": "600px",
                         "margin-right": "12px",
                         "margin-top": "24px",
                         "padding": "24px"})
            ,width= {"size":3}),
        dbc.Col([
            # This is the card for the pie chart
            dbc.Card(
                dbc.CardBody([
                    html.H5("Occurrence Map", style= {"text-align": "center"}),
                    html.Br([]),
                    dcc.Graph(
                        id="map",
                    )
                ]), style={
                           "height": "600px",
                           "margin-left": "12px",
                           "margin-right": "12px",
                           "margin-top": "24px",
                           }
            )],
            width=5),
        dbc.Col([
            # This is the card for the pie chart
            dbc.Card(
                dbc.CardBody([
                    html.H6("Simulated spatial distribution Ukranian shelf", style={"text-align": "center"}),
                    html.Img(
                        id = "img",
                        height=200,
                        style={"display": "block",
                               "margin-left": "auto",
                               "margin-right": "auto",
                               #"width": "50%",
                             }
                    ),
                    html.Br([]),
                    html.H6("HRMS/MS", style={"text-align": "center"}),
                    html.Div(
                        id="bar_chart",
                    )
                ]), style={
                    "height": "600px",
                    "margin-left": "12px",
                    "margin-top": "24px",
                }
            )],
            width=4),
    ]),
    html.Br([]),
    html.Br([]),
])


# #Create activity with Callbacks // The Output component will be changend depending on the input component
# @app.callback([
#     Output("source_checklist", "options"),
#     Output("source_checklist", "value")],
#     [Input("year_dropdown", "value")]
# )
# def table(year):
#     if year == "2016":
#         df = pd.read_excel("C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2016/Dataset_Y2016_map data.xlsx", sheet_name="Dataset_Y2016_map data")
#     elif year == "2017":
#         df = pd.read_excel("C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2017/Dataset_Y2017_map data.xlsx", sheet_name="Dataset_Y2017_map data")
#
#     values  = df["Predicted label"].unique()
#     options = [{"label": i, "value": i} for i in values]
#
#     return [options, values]


@app.callback([
    Output("compound_dropdown", "options"),
    Output("compound_dropdown", "value")],
    [Input("year_dropdown", "value"),
    Input("mz_input", "value"),
    Input("rti_input", "value"),
    Input("rt_input", "value"),
    Input("source_checklist", "value")]
)
def table(year, mz, rti, rt, sources):
    if year == "2016":
        df = df_1_16
    elif year == "2017":
        df = df_1_17

    df = df[df["Predicted label"].isin(sources)]

    if mz:
        df = df[df['Feature'].str.contains("mz=" + str(mz))]
    if rti:
        df = df[df['Feature'].str.contains("RTI=" + str(rti))]
    if rt:
        df = df[df['Feature'].str.contains("rt=" + str(rt))]

    options = [{"label": i, "value": i} for i in df["Feature"].unique()]
    value = df["Feature"].tolist()[0]

    return [options, value]


@app.callback(
    Output("map", "figure"),
    [Input("compound_dropdown", "value")],
    [State("year_dropdown", "value")], prevent_initial_call = True
)
def graph1(compound, year):

    if year == "2016":
        df_1 = df_1_16
        df_2 = df_2_16
    elif year == "2017":
        df_1 = df_1_17
        df_2 = df_2_17

    filtered_df_1 = df_1[df_1["Feature"] == compound]
    magnitude = filtered_df_1.values[0]
    df_2["Magnitude"] = magnitude[2:]
    fig = px.density_mapbox(df_2, lat='Latitude', lon='Longitude', z='Magnitude', radius=10,
                            center=dict(lat=43.5, lon=34), zoom=4,
                            mapbox_style="stamen-terrain")


    # fig = go.Figure(go.Densitymapbox(lat=df_2.Latitude, lon=df_2.Longitude, z=df_2.Magnitude))#, radius=10))
    # fig.update_layout(mapbox_style="stamen-terrain", mapbox_center_lon=34, mapbox_center_lat = 43.5, mapbox_zoom = 4, showlegend=False)

    fig.update_layout(margin={"t": 0, "b": 0, "l": 24, "r": 24},

                      # legend=dict(
                      #     orientation="h",
                      #     yanchor="bottom",
                      #     y=1.02,
                      #     xanchor="right",
                      #     x=1
                      # )
                      )

    return fig


@app.callback(
    Output("img", "src"),
    [Input("compound_dropdown", "value")],
    [State("year_dropdown", "value")], prevent_initial_call = True
)
def graph1(compound, year):
    if year == "2016":
        path = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2016/Simulated spatial distribution/" + compound + ".png"
    elif year == "2017":
        path = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2017/Simulated spatial distribution/" + compound + ".png"

    img_base64 = base64.b64encode(open(path, 'rb').read()).decode('ascii')
    src = 'data:image/png;base64,{}'.format(img_base64)

    return src


@app.callback(
    Output("bar_chart", "children"),
    [Input("compound_dropdown", "value")],
    [State("year_dropdown", "value")], prevent_initial_call = True
)
def graph1(compound, year):
    if year == "2016":
        path = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2016/HRMSMS/" + compound + ".csv"
    elif year == "2017":
        path = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2017/HRMSMS/" + compound + ".csv"

    try:
        df = pd.read_csv(path)

    except:
        return html.Div([html.Br([]), html.Br([]), html.Br([]), html.Br([]), html.Br([]), html.Br([]), html.Div("No data for this compound", style={"text-align": "center"})])

    fig = go.Figure([go.Bar(x=df.mz, y=df.rt)])
    fig.update_traces(marker_color='Grey', marker_line_color='Black') #,marker_line_width=1.5, opacity=0.6)
    fig.update_layout(plot_bgcolor='white',
                      height= 300,
                      margin={"t": 20, "b": 0, "l": 50, "r": 50},
                      legend={'font': {'family': "arial"}},
                      )

    fig.update_layout(legend=dict(
        font={'family': 'arial', 'size': 14, "color": "black"},
    ))

    fig.update_xaxes(title={'text': "Mass to charge ratio (m/z)", 'font': {'family': 'arial', 'size': 14, "color": "black"}},
                     showline=True, tickfont={'family': 'arial', 'size': 12, "color": "black"},
                     linewidth=1, linecolor='black', showgrid=False, mirror=True)
    fig.update_yaxes(title={'text': "Intensity (counts)", 'font': {'family': 'arial', 'size': 14, "color": "black"}},
                     showline=True, showgrid=False, gridwidth=0.5,gridcolor='lightgrey',
                     linewidth=1, linecolor='black', ticks='outside', tickfont={'family': 'arial', 'size': 12, "color": "black"},
                     mirror=True, rangemode="tozero")

    return html.Div([
                dbc.Button("Download CSV", id="btn_csv", style={"float": "right", "position": "relative", "z-index": "2", "margin-top": "25px", "margin-right": "60px"}),
                dcc.Download(id="download-dataframe-csv"),
                html.Br([]),html.Br([]),
                dcc.Graph(figure=fig,style={"margin-top": "-50px", "position": "relative", "z-index": "1"})
            ])


@app.callback(
    Output("download-dataframe-csv", "data"),
    Input("btn_csv", "n_clicks"),
    [State("compound_dropdown", "value"),
    State("year_dropdown", "value")], prevent_initial_call=True
)
def func(n_clicks, compound, year):
    if year == "2016":
        path = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2016/HRMSMS/" + compound + ".csv"
    elif year == "2017":
        path = "C:/Users/Nikiforos/Desktop/Dash/Black_sea/Dash task_02 Feb 2022/Y2017/HRMSMS/" + compound + ".csv"

    df = pd.read_csv(path)
    return dcc.send_data_frame(df.to_csv, "Y2016/HRMSMS/" + compound + ".csv", index=False)


#Starting the app
if __name__ == '__main__':
     app.run_server(debug=False)
```
